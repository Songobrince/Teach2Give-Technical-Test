```python
#Question 1: FizzBuzz
#Write a program that prints the numbers from 1 to 100. For multiples of 3, print "Fizz"; for
#multiples of 5, print "Buzz"; and for numbers that are multiples of both 3 and 5, print
#"FizzBuzz".


def fizz_Buzz():
   j=1
   while j < 100:
      
      if j%3==0 and j%5==0:
       print(f"{j} fizzbuzz")
      elif j%3==0:
       print(f"{j} fizz")
      elif j%5==0:
       print(f"{j} buzz")
      else:
       print(f"{j} not divisible ")

      j+=1

fizz_Buzz()
```
